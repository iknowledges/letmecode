```python
from typing import Optional
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def NodeToList(root: ListNode) -> list:
    result = []
    head = root
    while head:
        result.append(head.val)
        head = head.next
    return result
```

```python
def addTwoNumbers(l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
    carry = 0
    dump = ListNode()
    cur = dump
    while (l1 or l2):
        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0
        val = v1 + v2 + carry
        carry = 1 if val > 9 else 0
        cur.next = ListNode(val = val % 10)
        cur = cur.next
        print(f'carry={carry}, v1+v2+carry={val}, cur={NodeToList(cur)}, dump={NodeToList(dump)}')
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
    if carry > 0:
        cur.next = ListNode(val = carry)
    return dump.next
```

```python
# [2, 4, 3]
l1 = ListNode(2, ListNode(4, ListNode(3)))
# [5, 6, 4]
l2 = ListNode(5, ListNode(6, ListNode(4)))
output = [7,0,8]
# 342 + 465 = 807
assert NodeToList(addTwoNumbers(l1, l2)) == output
```
