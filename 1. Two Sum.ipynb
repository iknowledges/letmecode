```python
def twoSum(nums, target):
    d = dict()
    for i, num in enumerate(nums):
        diff = target - num
        if diff in d:
            print(f'num={num}, target-num={diff}, return dict={d}')
            return [d[diff], i]
        else:
            d[num] = i
            print(f'num={num}, target-num={diff}, update dict={d}')
    return []
```

```python
nums = [2,7,11,15]
target = 17
output = [0, 3]
assert twoSum(nums, target) == output
```
