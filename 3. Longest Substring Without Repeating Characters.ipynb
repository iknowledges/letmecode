```python
def lengthOfLongestSubstring(s: str) -> int:
    left = max_length = 0
    char_set = set()
    
    for right in range(len(s)):
        while s[right] in char_set:
            print(f'{s[right]} in {s[left: right+1]}, remove {s[left]}, left={left+1}')
            char_set.remove(s[left])
            left += 1

        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)
        print(f'current: {s[left: right+1]}, left={left}, right={right}, right-left+1={right - left + 1}, max_length={max_length}, char_set={char_set}\n')
    
    return max_length
```

```python
s = "abcabcbb"
output = 3
assert lengthOfLongestSubstring(s) == output
```
